```julia
using DynamicalSystems
using Plots
using Flux
using Flux: @epochs
using LinearAlgebra
# Graphs 
using LightGraphs, SimpleWeightedGraphs
using GraphPlot
using CUDA
using Optim
gr()
```

### Lorenz System

```julia
T = 2000
lorenz = DynamicalSystemsBase.Systems.lorenz([10.0, 10.0, 10.0];  = 10.0,  = 28.0,  = 8/3);
u = trajectory(lorenz, 100.0; dt = 0.02)
plot(u[1:T,1])
```

### Reservoir

```julia
D = 23
p = 1.2/D
G = erdos_renyi(D, p, is_directed=true)
while true
    
    G = erdos_renyi(D, p, is_directed=true)
    
    if is_connected(G) == true && maximum(abs.(eigvals(Array(adjacency_matrix(G))))) <1.3 && maximum(abs.(eigvals(Array(adjacency_matrix(G)))))>1.1
        break
    end
    
end
gplot(G, layout=circular_layout, nodelabel=1:D)
```

### Adjacency matrix

```julia
A = adjacency_matrix(G);
W = rand(0:100, D, 3)*0.0005;
maximum(abs.(eigvals(Array(A))))
```

### Equation of motion reservoir + driving

```julia
function loop(r, p, t)
    A = p[1]; W = p[2]
    r = []
    for i  1:D
        r = append!(r, tanh((Array(A[i,:])' * r) + (Array(W[i,:])' * u[t])))
    end
    return SVector{D}(r)
end
```

### Discrete dynamical system

```julia
p = [A, W]
Reservoir = DiscreteDynamicalSystem(loop, rand(D), p; t0 = 1);
```

```julia
tr = trajectory(Reservoir, T-1);
```

```julia
plot(tr[:,5])
```

### CUDA training

```julia
start_train = 10

X_train = CuArray(tr[start_train:end])
Y_train = CuArray(u[start_train:end])


 = Chain(Dense(D, 3))

loss(x,y) = sum(abs2, (x) .- y)
ps = Flux.params();

data = zip(X_train, Y_train);
#data = zip(tr[start_train:end], u[start_train:end]);
@time begin

    @epochs 1 Flux.train!(loss, ps, data, ADAM(0.001))

end
```

### Non CUDA training

```julia
start_train = 10

X_train = tr[start_train:end]
Y_train = u[start_train:end]


 = Chain(Dense(D, 3))

loss(x,y) = sum(abs2, (x) .- y)
ps = Flux.params();


data = zip(X_train, Y_train);
#data = zip(tr[start_train:end], u[start_train:end]);


#@time begin
#    @epochs 100 Flux.train!(loss, ps, data, ADAM(0.001))
#end

N = 5000
losses = []
for i  1:N
    Flux.train!(loss, ps, data, ADAM(0.001))
    
    l = 0
    for j  1:100
        l = l + loss(tr[2*j+10], u[2*j+10])
    end
    
    losses = push!(losses, l)
    if i%(5)==0
        println((i/N)*100, "%")
        println("loss: ", l)
    end
end
```

### Changing step size

```julia
N = 100
losses = []
for i  1:N
    Flux.train!(loss, ps, data, ADAM(0.01))
    
    l = 0
    for j  1:100
        l = l + loss(tr[2*j+10], u[2*j+10])
    end
    
    losses = push!(losses, l)
    if i%(5)==0
        println((i/N)*100, "%")
        println("loss: ", l)
    end
end
```

### Plot the loss
