```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.integrate import odeint
from sklearn.linear_model import Lasso
import tensorflow as tf
import keras
import wandb
import time
import pysindy as ps
```

```python
def lorenz(z, t):
    return [
        10 * (z[1] - z[0]),
        z[0] * (28 - z[2]) - z[1],
        z[0] * z[1] - (8 / 3) * z[2]
    ]
```

### Generate measurement data: $m=20$, $\tau = 0.05$

```python
dt = .05
N_train = 50
N_predict = 60

t_train = np.arange(0, dt*N_train, dt)
x0_train = [0.12849273,  0.35001826, 14.19480597] #starting point in the attractor
x_train = odeint(lorenz, x0_train, t_train)
```

## Generate the test data

```python
t_test = np.arange(0, (N_train+N_predict)*dt, dt)
x0_test = x0_train
x_test = odeint(lorenz, x0_test, t_test)
y_test = np.zeros((len(x_test), 1))
for i in range(t_test.size):
    y_test[i] = x_test[i][0]

plt.plot([i for i in range(len(x_train))], y_test[0:len(x_train)], marker = 'x', label = "training dataset")
plt.plot([i for i in range(len(x_train), len(x_test))], y_test[len(x_train):], marker = 'o', label = "to predict")
plt.legend()
plt.show()
```

## $h(\mathbf{x}(t)) = x_{1}(t)$

```python
y = np.zeros((len(x_train), 1))
for i in range(len(x_train)):
    y[i] = x_train[i][0]

x = x_train.astype(np.float32)[:-1]
```

```python
#wandb.init()

L = 60
y_predictions = y

start_time = time.time()

for i in range(L):
    y = y.astype(np.float32)[-len(x):]

    train_dataset = tf.data.Dataset.from_tensor_slices((x, y))
    train_dataset = train_dataset.batch(len(x))

    train_dataset.element_spec

    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(3, )),
        tf.keras.layers.Dense(128, activation='tanh'),
        tf.keras.layers.Dense(36, activation='tanh'),
        tf.keras.layers.Dense(1)
    ])

    model.compile(
        loss= tf.keras.losses.MeanSquaredError(),
        optimizer=keras.optimizers.Adam(0.001))
    
    history = model.fit(train_dataset, epochs=1500, verbose=0)
    
    model.compile(
        loss= tf.keras.losses.MeanSquaredError(),
        optimizer=keras.optimizers.Adam(0.0001))
    
    history1 = model.fit(train_dataset, epochs=500, verbose=0)

    y = np.append(y, [model(np.array([x_train[-1]]))[0][0].numpy()])
    
    y_predictions = np.append(y_predictions, [model(np.array([x_train[-1]]))[0][0].numpy()])
    
    print("predicted point: ", i)
    
print("running time : %s seconds" % (time.time() - start_time))
```

```python
plt.plot(y_test, marker='o', label = "true")
plt.vlines(N_train, -15, 10, colors='k', linestyle = 'dashed')
plt.plot([i for i in range(N_train, len(y_test))], y_predictions[N_train:], marker = 'x', label = "predicted")
plt.legend()
plt.show()
```
