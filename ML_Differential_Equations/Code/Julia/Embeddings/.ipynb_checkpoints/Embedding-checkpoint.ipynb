## Graph

```julia
N = 2
G = SimpleGraph(N)

while true

    for i ∈ 1:N
        for j ∈ i:N
            if i != j && (i == j +1 || i == j-1 || i == j-2 || i == j-2) #rand(Float64) > 0.9999
                add_edge!(G, i, j)
            end
        end
    end
    
    if is_connected(G) == true
        break
    end
end


#add_edge!(G, 1, 100)
#add_edge!(G, 1, 99)
#add_edge!(G, 2, 100)

#for i in 1:5
#    s = rand((1:100))
#    t = rand((1:100))
#    if s != t
#        add_edge!(G,s,t)
#    end
#end

gplot(G, layout=circular_layout, nodelabel=1:N)
```

## Adjacency matrix

```julia
A = adjacency_matrix(G);
```

## ODE

### Initial conditions

```julia
#Random initial conditions
#u0 = abs.(0.3*randn(Float32, (3*N,)));
u0 = [6*rand(Float32) for i in 1:3*N];
```

### System

```julia
function Rossler(u, A, t)
    
    K = 0.2
    
    S = [u[3*i-2] for i ∈ 1:N]    
    
    
    ODE = []
    
    for i in 1:N
        ODE = append!(ODE, [
            -u[3*i-1] - u[3*i] - (count(j->(j>0), Array(A[:, i])))*K*u[3*i-2] + K*S'*Array(A[:, i]),
            
            u[3*i-2] + 0.2*u[3*i-1],
            
            0.2 + u[3*i]*(u[3*i-2] - 5.7),
            
            ])
    end
    
    return ODE
end
```

```julia
tspan = (0.0f0, 50.0f0)
prob = ODEProblem(Rossler, u0, tspan, A)
saveat = 0.01
solution = solve(prob, Vern7(), saveat = saveat) #abstol=1e-12, reltol=1e-12,



X = Array(solution)

# Add noise to the data
#println("Generate noisy data")
Xₙ = X #+ Float32(1e-4)*randn(eltype(X), size(X));
```
