```julia
using DynamicalSystems, PyPlot

ds = Systems.gissinger(ones(3))
data = trajectory(ds, 1000.0, dt = 0.05);
```

```julia
xyz = columns(data)
plot(xyz[1], xyz[2])
```

```julia
figure(figsize = (12,10))
k = 1
for i in 1:3
    for τ in [5, 30, 100]
        R = embed(xyz[i], 2, τ)
        ax = subplot(3,3,k)
        plot(R[:, 1], R[:, 2], color = "C$(k-1)", lw = 0.8)
        title("var = $i, τ = $τ")
        global k+=1
    end
end

tight_layout()
suptitle("2D reconstructed space")
subplots_adjust(top=0.9)
```
